# üèÄ NBA Matchups (2007‚Äì2015) - Exploratory Data Analysis

This notebook provides a comprehensive exploratory data analysis (EDA) of NBA matchup data from 2007 to 2015.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configure plotting
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# Load all season data
file_paths = [
    "matchups-2007.csv", "matchups-2008.csv", "matchups-2009.csv",
    "matchups-2010.csv", "matchups-2011.csv", "matchups-2012.csv",
    "matchups-2013.csv", "matchups-2014.csv", "matchups-2015.csv"
]
dfs = [pd.read_csv(path) for path in file_paths]
df = pd.concat(dfs, ignore_index=True)
df.head()
```

```python
season_winrate = df.groupby('season')['outcome'].mean()
season_winrate.plot(marker='o')
plt.title("Home Win Rate Over Seasons")
plt.ylabel("Win Rate")
plt.xlabel("Season")
plt.grid(True)
plt.show()
```

## üë• Most Frequent Players in Matchups

```python
from collections import Counter

# Combine all player columns
player_cols = ['home_0', 'home_1', 'home_2', 'home_3', 'home_4',
               'away_0', 'away_1', 'away_2', 'away_3', 'away_4']
all_players = df[player_cols].values.ravel()
player_counts = Counter(all_players)
top_players = pd.DataFrame(player_counts.most_common(10), columns=['Player', 'Appearances'])

sns.barplot(x='Appearances', y='Player', data=top_players)
plt.title("Top 10 Most Frequent Players")
plt.show()
```

## üìà Team Performance Trends Over Time

```python
# Calculate win rate for each team by season
team_season_winrate = df.groupby(['season', 'home_team'])['outcome'].mean().reset_index()
pivot_table = team_season_winrate.pivot(index='season', columns='home_team', values='outcome')

# Plot top 5 teams with highest average win rate
top_teams = pivot_table.mean().sort_values(ascending=False).head(5).index
pivot_table[top_teams].plot(marker='o')
plt.title("Top 5 Home Teams Win Rate Over Seasons")
plt.ylabel("Win Rate")
plt.xlabel("Season")
plt.grid(True)
plt.legend(title='Team')
plt.show()
```

## üîó Player Co-occurrence Network

```python
import networkx as nx
from itertools import combinations

# Count co-occurrences of players in the same lineup (home or away)
cooccurrence = Counter()

for _, row in df.iterrows():
    for group in [row[['home_0','home_1','home_2','home_3','home_4']],
                  row[['away_0','away_1','away_2','away_3','away_4']]]:
        players = list(group.values)
        for pair in combinations(players, 2):
            cooccurrence[tuple(sorted(pair))] += 1

# Create network graph from top 50 most frequent co-occurrences
top_edges = dict(sorted(cooccurrence.items(), key=lambda x: x[1], reverse=True)[:50])
G = nx.Graph()
for (p1, p2), weight in top_edges.items():
    G.add_edge(p1, p2, weight=weight)

# Draw the network
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, k=0.5)
edges = G.edges(data=True)
weights = [edata['weight'] for _, _, edata in edges]

nx.draw(G, pos, with_labels=True, node_size=700, edge_color=weights, width=2.0, edge_cmap=plt.cm.Blues)
plt.title("Top 50 Player Co-occurrences in Lineups")
plt.show()
```

## üß© Interactive Team Performance Explorer

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Create dropdown for team selection
teams = sorted(df['home_team'].unique())
team_dropdown = widgets.Dropdown(
    options=teams,
    description='Team:',
    disabled=False
)

# Function to update plot
def plot_team_trend(team):
    clear_output(wait=True)
    display(team_dropdown)
    team_data = df[df['home_team'] == team]
    season_winrate = team_data.groupby('season')['outcome'].mean()
    season_winrate.plot(marker='o')
    plt.title(f"{team} Home Win Rate Over Seasons")
    plt.ylabel("Win Rate")
    plt.xlabel("Season")
    plt.ylim(0, 1)
    plt.grid(True)
    plt.show()

# Display widget
team_dropdown.observe(lambda change: plot_team_trend(change.new) if change.type == 'change' else None, names='value')
display(team_dropdown)
plot_team_trend(teams[0])
```

```python
# === Player List for Filter ===
player_cols = ['home_0', 'home_1', 'home_2', 'home_3', 'home_4',
               'away_0', 'away_1', 'away_2', 'away_3', 'away_4']
all_players = df[player_cols].values.ravel()
player_counts = Counter(all_players)
top_players = [player for player, _ in player_counts.most_common(20)]

# === Widgets ===
season_slider = widgets.IntRangeSlider(
    value=[2007, 2015],
    min=2007,
    max=2015,
    step=1,
    description='Seasons:',
    continuous_update=False
)

team_dropdown = widgets.Dropdown(
    options=sorted(df['home_team'].unique()),
    description='Team:',
)

player_dropdown = widgets.Dropdown(
    options=top_players,
    description='Player:',
)

stat_dropdown = widgets.Dropdown(
    options=['pts', 'reb', 'ast', 'to', 'pf', 'blk'],
    value='pts',
    description='Stat:',
)

homeaway_toggle = widgets.ToggleButtons(
    options=[('Home', 'home'), ('Away', 'away')],
    description='Side:',
    button_style=''
)

# === Callback ===
def update_plot(*args):
    clear_output(wait=True)
    display(season_slider, team_dropdown, player_dropdown, stat_dropdown, homeaway_toggle)

    # Filters
    season_min, season_max = season_slider.value
    team = team_dropdown.value
    player = player_dropdown.value
    stat = stat_dropdown.value
    side = homeaway_toggle.value

    # Filter season and team
    dff = df[(df['season'] >= season_min) & (df['season'] <= season_max)]
    dff = dff[dff[f'{side}_team'] == team]

    # Filter player
    player_cols_side = [f'{side}_{i}' for i in range(5)]
    dff = dff[dff[player_cols_side].isin([player]).any(axis=1)]

    # Group and Plot
    grouped = dff.groupby('season')[f'{stat}_{side}'].mean()
    grouped.plot(marker='o')
    plt.title(f"{team} ({side.title()}) Average {stat.upper()} with {player} ({season_min}-{season_max})")
    plt.ylabel(stat.upper())
    plt.xlabel("Season")
    plt.grid(True)
    plt.show()

# === Bind Events ===
season_slider.observe(update_plot, names='value')
team_dropdown.observe(update_plot, names='value')
player_dropdown.observe(update_plot, names='value')
stat_dropdown.observe(update_plot, names='value')
homeaway_toggle.observe(update_plot, names='value')

# === Display ===
display(season_slider, team_dropdown, player_dropdown, stat_dropdown, homeaway_toggle)
update_plot()
```
